```python
fig, axes= plt.subplots(figsize=(18,10))
axes.scatter(data["Ingresos Anuales"], data["Gasto Mensual"])
axes.set_xlabel("Ingresos Anuales")
axes.set_ylabel("Gasto Mensual")
plt.show()
```

```python
fig, axes= plt.subplots(figsize=(18,10))
axes.scatter(data["Ingresos Anuales"], data["Puntuación de Crédito"])
axes.set_xlabel("Ingresos Anuales")
axes.set_ylabel("Puntuación de Crédito")
plt.show()
```
